## Homework
Create and expand the Cookie Clicker project:
1. Fill out the cookies and cookiesPerClick variables.
2. Define what should happen upon clicking the cookie.
3. Create an `Upgrade` class that multiplies cookies per click, and expand the original cookieclicker class to integrate upgardes.
5. Print how each upgrade changes the total cookie output.
6. Add a cookie type variable which sets the specific type of cookie (ex: Chocolate chip, Oatmeal, etc.)
The following code is to help you get started.

#### Extra credit: Up to 0.03 points
- Create a new cell, apply the ALL of the above changes to a blank cookie clicker project, and submit that code for the cookie clicker project.

```python
%%javascript

class CookieClicker {
  constructor(cookies,cookiesPerClick) {
    this.cookies = 0; //start value
    this.cookiesPerClick = 1;
  }

  click() {
    this.cookies += this.cookiesPerClick;
    console.log(`You have ${this.cookies} cookies.`);
  }
}

class Grandma extends CookieClicker {
  constructor(cookies, cookiesPerClick, grandmas) {
    super(cookies, cookiesPerClick);
    this.grandmas = 0; //start value
  }

  buyGrandma() {
    if (this.cookies >= 10) {
      this.cookies -= 10;
      this.grandmas += 1;
      this.cookiesPerClick += 1; //each grandma adds 1 cookie per click
      console.log(`You bought a grandma! You now have ${this.grandmas} grandmas.`);
    } else {
      console.log("Not enough cookies to buy a grandma.");
    }
  }
}

class Farm extends CookieClicker {
  constructor(cookies, cookiesPerClick, farms) {
    super(cookies, cookiesPerClick);
    this.farms = 0; //start value
  }

  buyFarm() {
    if (this.cookies >= 100) {
      this.cookies -= 100;
      this.farms += 1;
      this.cookiesPerClick += 10; //each farm adds 10 cookies per click
      console.log(`You bought a farm! You now have ${this.farms} farms.`);
    } else {
      console.log("Not enough cookies to buy a farm.");
    }
  }
}

class Mine extends CookieClicker {
  constructor(cookies, cookiesPerClick, mines) {
    super(cookies, cookiesPerClick);
    this.mines = 0; //start value
  }

  buyMine() {
    if (this.cookies >= 1000) {
      this.cookies -= 1000;
      this.mines += 1;
      this.cookiesPerClick += 100; //each mine adds 100 cookies per click
      console.log(`You bought a mine! You now have ${this.mines} mines.`);
    } else {
      console.log("Not enough cookies to buy a mine.");
    }
  }
}

// Example usage:
const game = new CookieClicker();
game.click(); // You have 1 cookies.
game.click(); // You have 2 cookies.

const grandmaGame = new Grandma();
grandmaGame.click(); // You have 1 cookies.
grandmaGame.click(); // You have 2 cookies.
grandmaGame.buyGrandma(); // Not enough cookies to buy a grandma.
for (let i = 0; i < 10; i++) {
  grandmaGame.click();
}
grandmaGame.buyGrandma(); // You bought a grandma! You now have 1 grandmas.
grandmaGame.click(); // You have 4 cookies.

const farmGame = new Farm();
for (let i = 0; i < 100; i++) {
  farmGame.click();
}
farmGame.buyFarm(); // You bought a farm! You now have 1 farms.
farmGame.click(); // You have 11 cookies.

const mineGame = new Mine();
for (let i = 0; i < 1000; i++) {
  mineGame.click();
}
mineGame.buyMine(); // You bought a mine! You now have 1 mines.
mineGame.click(); // You have 101 cookies.  
```
